# Sentiment Analysis for Steam Reviews
Steam is a video game digital distribution service with a vast community of gamers globally. A lot of gamers write reviews at the game page and have an option of choosing whether they would recommend this game to others or not. However, determining this sentiment automatically from text can help Steam to automatically tag such reviews extracted from other forums across the internet and can help them better judge the popularity of games.

Given the review text with user recommendation and other information related to each game for 64 game titles, the task is to predict whether the reviewer recommended the game titles available in the test set on the basis of review text and other information.

Game overview information for both train and test are available in single file game_overview.csv inside train.zip

### Evaluation Metric
Submissions are evaluated on binary F1 Score between the predicted and observed user suggestion for the reviews in the test set.

### Text Processing
- Removing Punctuations 
- Removing Stop Words
- Lowercasing the Words
- Stemming the Words

```python
# Create a function to remove punctuations
def remove_punctuation(sentence: str) -> str:
    return sentence.translate(str.maketrans('', '', string.punctuation))

# Create a function to remove stop words
stop = stopwords.words('english')
def remove_stop_words(x):
    x = ' '.join([i for i in x.lower().split(' ') if i not in stop])
    return x

# Create a function to lowercase the words
def to_lower(x):
    return x.lower()
```

```python
# Apply Normalizing Functions on train data
df_train['user_review'] = df_train['user_review'].apply(remove_punctuation)
df_train['user_review'] = df_train['user_review'].apply(remove_stop_words)
df_train['user_review'] = df_train['user_review'].apply(to_lower)
df_train['title'] = df_train['title'].apply(remove_punctuation)
df_train['title'] = df_train['title'].apply(remove_stop_words)
df_train['title'] = df_train['title'].apply(to_lower)

# Apply Normalizing Functions on test data
df_test['user_review'] = df_test['user_review'].apply(remove_punctuation)
df_test['user_review'] = df_test['user_review'].apply(remove_stop_words)
df_test['user_review'] = df_test['user_review'].apply(to_lower)
df_test['title'] = df_test['title'].apply(remove_punctuation)
df_test['title'] = df_test['title'].apply(remove_stop_words)
df_test['title'] = df_test['title'].apply(to_lower)
```

```python
# set constant values
seed = 42
test_size = 0.30

# Create training and test sets
feature_train, feature_test, target_train, target_test = train_test_split(
    df_train["user_review"], 
    df_train.user_suggestion, 
    test_size=test_size, 
    random_state=seed)
```

```python
feature_train.shape, feature_test.shape
```

#### Count Vectors as features

```python
# Initialize a CountVectorizer object 
# transform the feature_train and feature_test data using count vectorizer object 
count_vectorizer = CountVectorizer(max_features = 50000) 
count_feature_train = count_vectorizer.fit_transform(feature_train)
count_feature_test = count_vectorizer.transform(feature_test)

# Print the first 10 features of the count_vectorizer
print(count_vectorizer.get_feature_names()[:10])
```

```python
len(count_vectorizer.get_feature_names())
```

#### TF-IDF Vectors as features

```python
tfidf_vectorizer = TfidfVectorizer(max_features=50000,max_df=0.7,ngram_range=(1, 2))
tfidf_feature_train = tfidf_vectorizer.fit_transform(feature_train)
tfidf_feature_test = tfidf_vectorizer.transform(feature_test)

# Print the first 10 features
print(tfidf_vectorizer.get_feature_names()[:10])

# Print the first 5 vectors of the tfidf training data
print(tfidf_feature_train.A[:5])
```

```python
def train_model(classifier, feature_vector_train, target_train_label, feature_vector_test, target_test_label):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, target_train_label)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_test)
    
    # Calculate the accuracy score:
    test_score = metrics.accuracy_score(target_test_label, predictions)
    
    # Calculate the confusion matrix: cm
    cf_matrix = metrics.confusion_matrix(target_test_label, predictions)
    
    # train cross validation score
    kf = KFold(n_splits=10, shuffle=True, random_state=42)
    train_cv_score = cross_val_score(classifier, feature_vector_train, target_train_label, scoring="f1", cv = kf).mean()
    
    # generate classification report
    model_clf_report = metrics.classification_report(target_test_label, classifier.predict(feature_vector_test))
    
    return train_cv_score, test_score, cf_matrix, classifier, model_clf_report
```

```python
# Instantiate a Multinomial Naive Bayes classifier and use Count Vectors features
nb_classifier =naive_bayes.MultinomialNB()
cv_score, test_score, cf_matrix, model, model_clf_report = train_model(nb_classifier, count_feature_train, target_train, count_feature_test, target_test)
print('Count Vectors')
print('------model scores------')
print('cv_score: ',cv_score)
print('test_score: ',test_score)
print(' ')
print('------confusion matrix------')
print(cf_matrix)
print(' ')
print('-----classification report-----')
print(model_clf_report)

print('')
print('')
nb_classifier =naive_bayes.MultinomialNB()
cv_score, test_score, cf_matrix, model, model_clf_report = train_model(nb_classifier, tfidf_feature_train, target_train, tfidf_feature_test, target_test)
print('TF-IDF Vectors')
print('------model scores------')
print('cv_score: ',cv_score)
print('test_score: ',test_score)
print(' ')
print('------confusion matrix------')
print(cf_matrix)
print(' ')
print('-----classification report-----')
print(model_clf_report)


# Instantiate a LogisticRegression classifier and use Count Vectors features
lr_classifier = linear_model.LogisticRegression()
cv_score, test_score, cf_matrix, model, model_clf_report = train_model(lr_classifier, count_feature_train, target_train, count_feature_test, target_test)
print('LogisticRegression')
print('Count Vectors')
print('------model scores------')
print('cv_score: ',cv_score)
print('test_score: ',test_score)
print(' ')
print('------confusion matrix------')
print(cf_matrix)
print(' ')
print('-----classification report-----')
print(model_clf_report)

print('')
print('')
lr_classifier = linear_model.LogisticRegression()
cv_score, test_score, cf_matrix, model, model_clf_report = train_model(lr_classifier, tfidf_feature_train, target_train, tfidf_feature_test, target_test)
print('TF-IDF Vectors')
print('------model scores------')
print('cv_score: ',cv_score)
print('test_score: ',test_score)
print(' ')
print('------confusion matrix------')
print(cf_matrix)
print(' ')
print('-----classification report-----')
print(model_clf_report)
```
